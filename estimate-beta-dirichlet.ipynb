```python
'''
ex: K = 10; N = 52;
ns = {n_l:l=1,...K} = [0,30,1,2,0,0,9,6,3,1] -> beta
'''

def Derivative(ns, beta):
    k=len(ns)
    N=sum(ns) 
    #print(k,N)
    s1=0
    for n in ns:
        for m in range(n):
            d=m+beta
            s1+=1./d
    s2=0
    for n in range(N):
        d=n+k*beta
        s2+=k/d
    q=s1-s2
    return q

def DerivativeNSB(ns, beta):
    k=len(ns)
    N=sum(ns) 
    #print(k,N)
    s1=0
    for n in ns:
        for m in range(n):
            d=m+beta
            s1+=1./d
    s2=0
    for n in range(N):
        d=n+k*beta
        s2+=k/d
    q=s1-s2
    d2sbar=k*k*polygamma(2,k*beta+1)-polygamma(2,beta+1)
    dsbar=k*polygamma(1,k*beta+1)-polygamma(1,beta+1)
    #print(beta,q,d2sbar)
    return q+d2sbar/dsbar

def beta_star(ns, niter=10): 
    beta=[0.00000001,0.0000001,0.000001,0.00001,0.0001,0.001,0.01,0.1,1,10,100,1000,10000,100000,1000000,10000000]
    for iter in range (niter):
        if iter==0: betas=beta
        b=betas[0]
        #print(b)    
        q=Derivative(ns,b)
        s=-1
        if q>0: s=1
        #print(iter,q,s)
        for b in betas:
            q=Derivative(ns,b)
            #print(b,q)
            if q*s<0:
            #change of sign
                if iter==0:
                    betas=[b/10+n*b/10 for n in range(11)]
                    #print('i0 change',b)
                    break
                else:
                    db=betas[1]-betas[0]
                    betas=[b-db+db/10*n for n in range(11)]
                    #print('change',b,q)
                    break

    # If no solution was found, select among extremes
    if b == beta[-1]:
        if Derivative(ns, b) < 0:
            b = beta[0]
    # Done
    return b, Derivative(ns,b)

def beta_starNSB(ns, niter=10):
    beta=[0.00000001,0.0000001,0.000001,0.00001,0.0001,0.001,0.01,0.1,1,10,100,1000,10000,100000,1000000,10000000]
    for iter in range (niter):
        if iter==0: betas=beta
        b=betas[0]
        #print(b)    
        q=DerivativeNSB(ns,b)
        s=-1
        if q>0: s=1
        #print('iter',iter,q,s)
        for b in betas:
            q=DerivativeNSB(ns,b)
            #print(b,q)
            if q*s<0:
            #change of sign
                if iter==0:
                    betas=[b/10+n*b/10 for n in range(11)]
                    #print('i0 change',b)
                    break
                else:
                    db=betas[1]-betas[0]
                    betas=[b-db+db/10*n for n in range(11)]
                    #print('change',b,q)
                    break

    # If no solution was found, select among extremes
    if b == beta[-1]:
        if DerivativeNSB(ns, b) < 0:
            b = beta[0]
    # Done
    return b, DerivativeNSB(ns,b)
```
